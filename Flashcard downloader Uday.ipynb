```python
import requests
import sys
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = 'http://www.dbtventures.com/data-science'
result = requests.get(url)
```

```python
c = result.content
```

```python
soup = BeautifulSoup(c)
imgs = soup.findAll('div', attrs={'class': 'margin-wrapper'})
dic = {}
```

```python
names = []
image_urls = []
for image in imgs:
    image_url = image.img["src"]
    name = re.sub(re.compile('_web.png'),"",image.img["alt"])
    names.append(name)
    image_urls.append(image_url)
```

```python
d = {'Image_name': names, 'Image_url': image_urls}
df = pd.DataFrame(data=d)
```

```python
df
```

```python
df.to_csv('ML_flashcards.csv')
```

```python
import urllib.request
for i in range (len(names)):
    urllib.request.urlretrieve(df["Image_url"][i], names[i]+".png")
```
