```
id_zip = data[['School_ID','Zip']]
```

```
school_counts = id_zip.groupby('Zip', as_index=False).count()
school_counts = school_counts.rename(columns={'School_ID':'school_counts'})
school_counts
```

```
import plotly.graph_objects as go
import plotly.express as px
data = [go.Bar(x=list(school_counts.Zip), y=school_counts['school_counts'], 
               text = 'school_counts', hovertext=['school_counts'],
               marker={'color': school_counts['school_counts'], 'colorscale': 'Viridis', "showscale": True}, 
               )]
layout = go.Layout(xaxis=dict(type='category'))
fig = go.Figure(data=data, layout=layout)
fig.update_xaxes(
        tickangle = 90,
        title_text = "Zip",)
fig.update_yaxes(
        title_text = "Number of Schools")
fig.update_layout(title="Number of Schools per zipcode", hovermode='x')
fig.show()
```

## Choropleth map

```
!pip install geopandas
```

```
import folium
import geopandas

df = geopandas.read_file("Boundaries - ZIP Codes.geojson")
df
```

```
# https://github.com/python-visualization/folium/issues/917#issuecomment-406879901
school_counts['Zip']=school_counts['Zip'].astype(str)
```

```
school_counts.info()
```

```
#definition of the boundaries in the map
Chicago_COORDINATES = (41.895140898, -87.624255632)
district_geo = r'Boundaries - ZIP Codes.geojson'

# https://data.cityofchicago.org/Facilities-Geographic-Boundaries/Boundaries-Wards-2015-/sp34-6z76
#creating choropleth map for Chicago District 2016
map1 = folium.Map(location=Chicago_COORDINATES, zoom_start=11)
folium.Choropleth(geo_data = district_geo,
                data = school_counts,
                columns = ['Zip', 'school_counts'],
                key_on = 'feature.properties.zip',
                fill_color = "Blues",
                legend_name = 'Number of schools per zip code').add_to(map1)

map1
```
